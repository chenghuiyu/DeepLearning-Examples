# linear regression


```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

At the beginning, we can compute the following equation: $Y = WX + b$, where $W$ and $X$ are random matrices and b is a random vector.

```python
# Parameters
learning_rate = 0.01
training_epochs = 1000
display_step = 25
```

```python
# Training Data
train_X = np.asarray([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,
                         7.042,10.791,5.313,7.997,5.654,9.27,3.1])
train_Y = np.asarray([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,
                         2.827,3.465,1.65,2.904,2.42,2.94,1.3])
n_samples = train_X.shape[0]
```

```python
# define tf Graph Input
X = tf.placeholder(tf.float32, name="X")
Y = tf.placeholder(tf.float32, name="Y")

# set model weights and bais
W = tf.Variable(np.random.randn(), name = "weight")
b = tf.Variable(np.random.randn(), name = "bais")
```

Then we will use equation: $Y = WX + b$ to construct a linear model

```python
Y_pre = tf.add(tf.multiply(X, W), b)
```

We can use several Algorithms to compute cost

$$ J = - \frac{1}{m}  \sum_{i = 1}^m  \large ( \small y^{(i)} \log a^{ [2] (i)} + (1-y^{(i)})\log (1-a^{ [2] (i)} )\large )\small\tag{2}$$


```python
logits = tf.sigmoid(Y_pre)
cost_coss = tf.nn.sigmoid_cross_entropy_with_logits(logits=logits, labels=Y)
```

we can use mean squared error to compute cost
$$ cost = \frac{1}{2*n}  \sum_{i = 1}^m  ( \small y^{(i)} - y)^{2}\small\tag{2}$$

```python
cost_mean = tf.reduce_mean(tf.pow((Y_pre - Y), 2))/(2 * n_samples)
```

```python
# Gradient descent, we use cost_mean for example
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost_mean)
```

```python
# Initialize the variables (i.e. assign their default value)
init = tf.global_variables_initializer()
```

```python
# start training
with tf.Session() as sess:
    sess.run(init)
    
    # feed all train data
    for epoch in range(training_epochs):
        for (x, y) in zip(train_X, train_Y):
            sess.run(optimizer, feed_dict={X:x, Y:y})
            
            # display logs per epoch step
            if((epoch + 1) % display_step == 0):
                c = sess.run(cost_mean, feed_dict={X:train_X, Y:train_Y})
                print "Epoch:", "%04d" % (epoch + 1), "cost = ", "{:.9f}".format(c),\
                      "W = ", sess.run(W), "b = ", sess.run(b)
                    
    print "Optimization Finished!"
    train_cost = sess.run(cost_mean, feed_dict={X:train_X, Y:train_Y})
    print "train_cost = ", train_cost, "W = ", sess.run(W), "b = ", sess.run(b), '\n'
    
    #Graphic display
    plt.plot(train_X, train_Y, 'ro', label = 'train data')
    plt.plot(train_X, sess.run(W) * train_X + sess.run(b), label = 'fitted line')
    plt.legend()
    plt.show()
```
